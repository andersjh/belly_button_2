```python
import pandas as pd
import json
from pprint import pprint
import re
```

```python
# testing regular expression splitting with possible spaces
rex = re.compile("\s*[/,]\s*")
rex.split("Carlotte /NC")
```

```python
with open("samples.json", "r") as fh:
    raw = fh.read()
    data = json.loads(raw)
    pprint(data)

#     print(data.keys())
#     print('-' * 25, "names")
#     print(data['names'], len(data['names']))
#     print('-' * 25, "metadata")
#     print(data['metadata'], len(data['metadata']))
#     print('-' * 25, "samples")
#     print(data['samples'], len(data['samples']['otu_ids']))
#     print('-' * 25)
```

```python
df_people = pd.DataFrame(data['metadata'])
df_people['city'] = ''
df_people['state'] = ''
for k, v in df_people.iterrows():
    val = str(v.location)
#     print(val)
#     print(type(v.location))
    vals = rex.split(val)
#     print(vals)
    if len(vals) == 2:
        df_people.loc[k, 'city'] = vals[0]
        df_people.loc[k, 'state'] = vals[1]  
    else:
        df_people.loc[k, 'state'] = val
    
df_people.to_csv("subjects.csv", index=False)
df_people.head()
```

```python
otu_dict = {}
test_data = []
samples = data['samples']
for sample in samples:
#     pprint(sample)
    cur_id = int(sample['id'])
    otu_ids = sample['otu_ids']
    otu_labels = sample['otu_labels']
    sample_values = sample['sample_values']
#     print(cur_id, len(otu_ids), len(otu_labels), len(sample_values))
    for i in range(len(otu_ids)):
        otu_dict[otu_ids[i]] = otu_labels[i]
        test_data.append(
            {
                "subject_id": cur_id,
                "original_sequence": i,
                "otu_id": otu_ids[i],
                "amount": int(sample_values[i])
            }
        )
print("all done")
        
    
```

```python
otu = pd.DataFrame.from_dict(otu_dict, orient='index', columns=['name'])
otu.reset_index(inplace=True)
otu.columns = ['id', 'name']
otu.head()
```

```python
otu.to_csv("otu_reference.csv", index=False)
print("file saved")
```

```python
td = pd.DataFrame(test_data)
td.to_csv("test_data.csv", index=False)
print("\nfile saved\n")
td.head()
```
