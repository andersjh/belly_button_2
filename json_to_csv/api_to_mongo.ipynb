```python
otu_dict = {}
test_data = []
samples = data['samples']
for sample in samples:
#     pprint(sample)
    cur_id = int(sample['id'])
    otu_ids = sample['otu_ids']
    otu_labels = sample['otu_labels']
    sample_values = sample['sample_values']
#     print(cur_id, len(otu_ids), len(otu_labels), len(sample_values))
    for i in range(len(otu_ids)):
        otu_dict[otu_ids[i]] = otu_labels[i]
        test_data.append(
            {
                "subject_id": cur_id,
                "original_sequence": i,
                "otu_id": otu_ids[i],
                "amount": int(sample_values[i])
            }
        )
print("all done")
        
    
```

```python
otu = pd.DataFrame.from_dict(otu_dict, orient='index', columns=['name'])
otu.reset_index(inplace=True)
otu.columns = ['id', 'name']
otu.head()
```

```python
otu.to_csv("otu_reference.csv", index=False)
print("file saved")
```
